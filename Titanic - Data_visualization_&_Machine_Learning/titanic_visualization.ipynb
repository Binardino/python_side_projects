**Data Wranling**

2 columns have redondant data :
- SibSp - The number of siblings or spouses the passenger had aboard the Titanic
- Parch - The number of parents or children the passenger had aboard the Titanic

Adding up two columns in one global Relative_Nb colum

```python
df_titanic['Relative_Nb'] = df_titanic['SibSp'] + df_titanic['Parch']
```

```python
df_titanic
```

First thing first : let's transform all categorial values to dummy variables

```python
#Transforming Gender categorical into dummy variable
df_titanic_dummy = pd.get_dummies(data=df_titanic, columns=['Gender'], prefix='Gender')

df_titanic_dummy
```

### Selecting relevant columns

Several column info are useless for the ML prediction = dropped :
- cabin number (Ticket Class more relevant)
- Embarked City (may be an indicator of richness for each passenger, when checking which city was the wealthiest in average. But once again Ticket Class & Ticket Fare are more straightforward richness indicator)
- Ticket Number : just alphanumc mix, no meaning
- droping PassengerId
- Keeping only column of the One Hot encoded Gender column - Gender female : hereafter 1 for women / 0 for men 

```python
df_titanic_clean = df_titanic_dummy[['PassengerId', 'Survived', 'Pclass', 'Gender_female', 'Age', 'Relative_Nb', 'Fare']].copy()
```

```python
df_titanic_clean.rename(columns={'Gender_female':'Gender'}, inplace=True)
```

**What does the last plot tell you about the feature Age?**
The boxenplot displays the several Quartile of the Age distribution :

- the min & max values
- the median (highlighted by the blue dashed axvline)
- the Q1 (25%) & Q3 (75%) quartiles

**Conclusion**

50% of the population between 23 & 37 years old

# MatplotLib global Plots

```python
survivors = df_titanic[df_titanic['Survived'] == True]

non_survivors = df_titanic[df_titanic['Survived'] == False]

#creating figure
figure = plt.figure(figsize=(20,10))
figure.suptitle('Survivors VS. Non-Survivors', fontsize=16)

#creating grid
plt.rc('axes', facecolor='#ececec', edgecolor='none', axisbelow=True, grid=True)
plt.rc('grid', color='w', linestyle='solid')

#adding space btw subplots
plt.subplots_adjust(top=0.9, bottom=0.1, hspace=0.3, wspace=0.4)

#Pclass (#1)
plt.subplot(2,4,1)
survivors_data = survivors.groupby('Pclass').size().values
non_survivors_data = non_survivors.groupby('Pclass').size().values
plt.bar(range(len(survivors_data)), survivors_data, label='Survivors', alpha=0.5, color='g')
plt.bar(range(len(non_survivors_data)), non_survivors_data, label='non_Survivors', alpha=0.5, color='r')
plt.xticks([0,1,2],[1,2,3])
plt.ylabel('Count')
plt.title('Pclass')
plt.legend(loc='upper left')

#Plot by Gender #2
plt.subplot(2,4,2)
survivors_data = survivors.groupby('Gender').size().values
non_survivors_data = non_survivors.groupby('Gender').size().values
plt.bar(range(len(survivors_data)), survivors_data, label='Survivors',color='g')
plt.bar(range(len(non_survivors_data)), non_survivors_data, label='non-Survivors',color='r')
plt.xticks([0,1],['Female', 'Male'])
plt.ylabel('Count')
plt.title('Gender')
plt.legend(loc='upper left')

#Plot by Age #3
plt.subplot(2,4,3)
survivors_data = survivors.groupby('Age').size().values
non_survivors_data = non_survivors.groupby('Age').size().values
plt.bar(range(len(survivors_data)), survivors_data, label='Age', color='g')
plt.bar(range(len(non_survivors_data)), non_survivors_data, label='Age', color='r')
plt.xticks([0,1,2,3,4,5,6,7], ['0-9', '10-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70-79'], rotation=40, horizontalalignment='right')
plt.ylabel('Count')
plt.title('Age')
plt.legend(loc='upper left')
```
